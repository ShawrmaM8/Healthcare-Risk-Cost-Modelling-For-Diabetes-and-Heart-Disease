## 1.1 Load & Inspect

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hd = pd.read_csv(r"datasets\heart.csv")
hd_df = pd.DataFrame(hd)
hd_df
```

#### Clarification on meaning of each column metric:

| Feature      | Meaning (Simple)                                                                                                                           |
| ------------ | ------------------------------------------------------------------------------------------------------------------------------------------ |
| **age**      | Patient's age in years                                                                                                                     |
| **sex**      | 1 = male, 0 = female                                                                                                                       |
| **cp**       | Type of chest pain:<br>0 = typical angina (heart-related)<br>1 = atypical angina<br>2 = non-anginal pain<br>3 = no pain but other symptoms |
| **trestbps** | Resting blood pressure (in mm Hg)                                                                                                          |
| **chol**     | Cholesterol level (mg/dL)                                                                                                                  |
| **fbs**      | Fasting blood sugar > 120 mg/dL:<br>1 = Yes, 0 = No                                                                                        |
| **restecg**  | Resting ECG (heart rhythm test) results:<br>0 = normal<br>1 = some abnormality<br>2 = definite abnormality                                 |
| **thalach**  | Max heart rate achieved during exercise                                                                                                    |
| **exang**    | Did exercise cause chest pain? 1 = Yes, 0 = No                                                                                             |
| **slope**    | Slope of heart's response to exercise:<br>0 = flat<br>1 = up (healthy)<br>2 = down (problematic)                                           |
| **ca**       | Number of major blood vessels (0–3) seen in scan                                                                                           |
| **thal**     | Type of thalassemia (blood flow issue):<br>0 = normal<br>1 = fixed defect<br>2 = reversible defect                                         |
| **target**   | 1 = Has heart disease, 0 = No heart disease                                                                                               |
           |

### Conclusion:

#### 1. Class Imbalance:

* Gender (sex) is not evenly distributed between both classes
* Overwhelming majority of samples collected are from male participants (69.6%) compared to female participants (39.4%)

#### 2. Machine-Learning Implications:

##### **Bias:**
Training dataset will be heavily biased due to uneven representation of both Gender classes, reducing the reliability of the ML model

##### There are 2 most efficient ways to handle this:
1. **Resampling:**
* Undersampling to reduce male samples
* Oversampling to increase female samples *(i.e. Synthesizing using SMOTE)*
  
2. **Class weights:**
* Assign higher weights to underrepresented samples *(i.e. `LogisticRegression(class_weight="balanced")`)*
* This allows for equal treatment of the two classes during Model training

#### 3. Real-World Implications

* 51.3% heart disease rate is extremely high relative to global prevalence (6.5%)
* This indicates predictions made from training data will be heavily skewed, leading to potentially false diagnosis.
Critical consequences likelihood: 8/10

## 1.3 Correlation Analysis

```python
num_df = hd_df.select_dtypes(include=["integer"]) # selects all numeric columns
num_df
```

```python
correlation = num_df.corr(method="pearson")
correlation
```

### Example summary:
* #### `0 correlation` between **cholestrol** *(chol)* and **type of chest pain** *(cp)*
  Cholestrol levels do not impact chest pain type at all
* #### **Weak negative correlation** `(-0.38)` between chest pain from exercise *(exang)* and max heart rate during exercise *(thalach)*
  Increased max heart rate during exercise (slightly) decreases chest pain from exercise
* #### **Weak positive correlation** `(0.346)` between **heart's response to exercise** *(slope)* and **heart disease diagnosis outcome** *(target)*
  Problematic heart response (2) from exercise is (slightly) likely to cause heart disease.
  On the other hand, normal heart response (0) from exercise hints less likelihood of heart disease

### Heatmap 

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data=correlation, annot=True, cbar=True, cmap="Greys")
plt.title("Heart Disease Correlation Heatmap")
plt.tight_layout()
plt.show()
```
